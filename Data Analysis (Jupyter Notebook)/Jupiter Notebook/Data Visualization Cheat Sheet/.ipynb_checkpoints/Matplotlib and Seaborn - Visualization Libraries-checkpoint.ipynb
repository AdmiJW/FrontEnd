### 2.1 Figures


`Matplotlib` operate on __figures__. All plotting is done on a `matplotlib` figure. Therefore we could change the properties of the figure using `plt.figure()` function.

We have to import `matplotlib` first. However, we are not using only `matplotlib` by itself, all of our functions are contained inside `matplotlib.pyplot`, so import that instead!

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python



# First, we could change the figure size easily, using keyword arguments. It takes in the width and height in inches.
plt.figure( figsize=(12,6) )


# Change the title of the figure by using .title()
plt.title( "Hello Plot!" )


# We could have subplots, which means in a figure there can be multiple plots in it.
plt.subplot( 1, 2 , 1 )            # What argument means:   ( No of rows, No of columns, Current selected plot )
plt.subplot( 1, 2, 2 )             # The numbering of plot is left to right, then new row
````

````python
# There are more advanced subplot2grid, which basically is like subplots in a grid system.
plt.figure( figsize=(14,6) )

# Arguments: ( shape of whole grid (3x3), starting index position, colspan=1, rowspan=1 )
ax1 = plt.subplot2grid( (3,3), (0,0), colspan=3 )
ax2 = plt.subplot2grid( (3,3), (1,0), colspan=2 )
ax3 = plt.subplot2grid( (3,3), (1,2), rowspan=2 )
ax4 = plt.subplot2grid( (3,3), (2,0) )
ax5 = plt.subplot2grid( (3,3), (2,1) )

plt.show()
````

### 2.2 Basic Plotting

Plotting a chart in `matplotlib` is as easy as just calling `plot()` function. Of course to create a nicer representation we can use some functions which alter each visualization element

Let's see some of the matplotlib functions to manipulate how the plot looks:
* `plt.title()` - set the title of the plot
* `plt.xlabel()` - set the x label of the graph
* `plt.ylabel()` - set the y label of the graph
* `plt.legend()` - add a legend to the graph. Either just pass the label in the plotting function or pass into this function argument


We may want to change the styling of fonts in the graph. To achieve that, some of the function may take in a `fontdict` argument, which is the properties of the font. Eg:
```python
fontdict = {
    'fontname': 'Comic Sans MS',
    'fontsize': 20,
    'fontweight': 300
}
```

When plotting line plots using `plot()`, it is common to use the `Line2D` shorthand arguments. Let's see how it is:

__Long, expanded version without shorthand:__
```python
plt.plot( x,y, label = '2x',
          color = 'red',
         linewidth = 2,
         marker = '.',
         linestyle = '--',
         markersize = 10,
         markeredgecolor = 'blue'
        )
```

The shorthand uses the format `[color][marker][line]`. For the possible selections, see `plot()` documentation.
```python
plt.plot( x, y, 'r^--', label='2x' )        # Line is red color, marker is a upper triangle, and dashed line
```


````python
# Here we plot a basic x^2 graph

x = np.arange( -10, 11 )     # [-10, -8, -6... 8, 10]

plt.figure( figsize=(8,6) )
plt.title(" Graph of X^2 ")

plt.plot(x, x**2 )    # Pass in the x values and y values

plt.xlabel('x')
plt.ylabel('y')

plt.legend(['x^2'] )

plt.show()

````

````python

# Let's combine 2 lines in one graph

x = np.arange(-10, 11 )

plt.figure( figsize=(8,6) )
plt.title( " My Combined Plot of -x^2" )

plt.xlabel("x")
plt.ylabel("y")


# This plots -x^2. The third argument 'rx--' is a shorthand for the styling. See on that later
plt.plot(x, x**2 * -1, 'rx--', label='-x^2' )

# This plots a vertical straight line in the middle of chart
plt.plot([0,0,0] , [0, -50, -100], label='Vertical Line')


plt.legend( )       # Show the legend based on the label argument in each plot()

plt.show()
````

````python

# There is no need to call the matplotlib each time we want to plot or manipulate our graphs. We could use the OOP way of doing
# it. We have to get the reference to the axes object in that case.

x = np.arange(-10, 11)

# Returns a 2D array. In each row consists of columns of individual plots!
plot_obj = plt.subplots( 2, 2, figsize=(14, 6) )

# We'll use destructuring syntax to get each individual plots
fig, ( (ax1, ax2), (ax3, ax4) ) = plot_obj

# As you can see, each plot is done by calling the plot objects, not matplotlib.plot()
ax1.plot( x, x**2, 'r^-' )
ax2.plot( x, x**2 * -1, 'bv-')
ax3.plot( x, x**3 , 'go--')
ax4.plot( x, x**3 * -1, c='orange', marker='o', linestyle=':')
````

### 2.3 - Scatterplots

So far we've been plotting line charts. Let's see how we can plot a scatterplot

````python

# To plot scatterplot, we need arrays of same size which consists of the x values and y values to plot
x = np.random.rand(50)
y = np.random.rand(50)

# Each dot will have different areas (size)
area = np.pi * ( np.random.rand(50) * 10 ) ** 2

# Colors will also be randomized
colors = np.random.rand(50)


plt.figure( figsize=(14,6) )

# Colors is a great way to represent a third dimension of data, other than x and y axis!
plt.scatter( x, y, s=area, c=colors, alpha=.6, cmap='Spectral' )

# Show the colorbar to show the values of each dot
plt.colorbar()


plt.show()
````

### 2.4 - Histogram

A histogram is kind of a bar chart, but the y value is the frequency / occurrences of the unique values or a 'bucket' of values in a particular column

It can be occurrences of individual values, like `male` or `female`.
It also can be occurrences of a range of values, seperated in 'buckets', like a data of ages of population can be put into buckets of range 10 like `11-20`, `21-30` etc...

````python

agesOfPeople = np.random.randint(low=1, high=101, size=1000)

plt.subplots( figsize=(12,6) )


# Plot the histogram
plt.hist(   agesOfPeople, 
            bins = 10,           # How many 'bins'. Putting 10 bins will be like 0-10, 11-20, 21-30...
            alpha=.5,
            histtype='bar',
             color='steelblue',
             edgecolor='blue'
        )

# plt.xlim( left=20, right=60 )    # 'Zoom in the data, so that we can see only for x values from 20 to 60'

plt.show()
````

### 2.6 - Pie Chart

A pie chart shows clearly the percentage of certain data groups in the column. It provides a direct, easy to understand way to visualize the data


````python

values = pd.Series( np.random.randint(1, 10001, size=10000) )

below20 = values[(values < 2000)].count()
above70 = values[(values > 7000)].count()
middle = values.count() - below20 - above70

# ======================================================

plt.style.use('ggplot')

pieData = [below20, above70, middle]
pieLabel = ['Below 2000', 'Above 7000', '2000 - 7000']

plt.pie( pieData, 
         labels=pieLabel, 
         autopct = '%.2f %%',               # Shows the percentage. We give it a formatting string
         pctdistance = 0.6,                 # Distance of the percentage label from the center of pie chart
         explode = (.01, .01, .01) )          # 'Explode' each segments of the pie chart from center. Higher value means more 
                                            # 'cracked'

plt.show()
````

### 2.7 - KDE (Kernel Density Estimation)

Kernal Density Estimation (KDE), with its intimidating name, is a way for us to quickly create a smooth curve out of a dataset. Instead of looking at discrete, solid shaped histograms, we can just look at KDE for its smooth curves.

To plot a KDE graph, we will need help from `scipy` library, which most of the scientific computation function resides. 

````python

from scipy import stats

values = np.random.randint(1, 1001, size=200 )
density = stats.kde.gaussian_kde( values )           # This is a class constructor. Later we will get individual values by
                                                     # putting in the x values, it will return the y value

    
# Create a numpy array linear spaced between the minimum and maximum value provided, at provided size
xValues = np.linspace( min(values) - 10, max(values) + 100, 100 )

# ------------ PLOTTING ---------------------

plt.plot( 
    xValues,
    density(xValues),
    color = 'green',
)


# It does plot out a line chart, but we don't quite see like KDE, which area under it shall be filled with color as well.
# We need to fill it by using matplotlib fill_between() function

plt.fill_between( 
    xValues,                         # X position to fill
    0,                               # Starting y position to fill
    density(xValues),                # Ending y position to fill
    alpha=0.5, 
    color='green' 
)

plt.show()
````

![Bar](https://github.com/AdmiJW/Items/blob/master/SeperatingBars/Horizontalbar_blue.png?raw=true)

### 2.8 - Composite Plots

A plot are not limited to only 1 type of graph! We can have 2 or more types of graph inside one subplot. The more popular plot combination are __Histogram__ and __KDE__ graph.

For compositing plots, we simply just call the plotting function on the same figure. It will plot on the same subplot normally

````python

from scipy import stats

agesOfPeople = np.random.randint(1,100, size=10000 )

density = stats.kde.gaussian_kde( agesOfPeople )

xValues = np.linspace(1, 100, 100)           # Or just np.arange(1,100)


#------------------------- PLOT -------------------------

plt.figure( figsize=(12,8) )

plt.plot( xValues, density(xValues), color='green')
plt.fill_between( xValues, 0, density(xValues), alpha=0.5, color='green')

plt.hist( agesOfPeople, bins=10, alpha=0.5, density=1, histtype='bar', color='steelblue', edgecolor='green', width=10 )

plt.xticks( np.linspace(5, 95, 10) )


plt.show()
````

# 3.0.a - Styling

![Bar](https://github.com/AdmiJW/Items/blob/master/SeperatingBars/Horizontalbar_green.png?raw=true)

We have some control over how the plot is drawn in `seaborn`. 

First of all, we can change the style of the plot. Like so:
```python
    sns.set_style('white')
```
The available options are:
* `white`
* `dark`
* `whitegrid`
* `darkgrid`
* `ticks`

````python
plt.figure( figsize=(16,5) )

sns.set_style('white')
plt.subplot(1,5,1 )

sns.set_style('dark')
plt.subplot(1,5,2)

sns.set_style('whitegrid')
plt.subplot(1,5,3)

sns.set_style('darkgrid')
plt.subplot(1,5,4)

sns.set_style('ticks')
plt.subplot(1,5,5)

plt.show()
````

Alongside with `set_style`, we can also use `set_context()`. In it we can set the context of the plot, to either:
* `paper` - Meant to be printed on paper
* `talk` - Meant to be used in presentations
* `poster` - Meant to be printed out on a poster

Also, we could use `font-scale` attribute to set the size of fonts

````python
plt.figure( figsize=(20, 8) )

plt.subplot(1,3,1)
plt.title('paper')
sns.set_context('paper', font_scale=1 )

plt.subplot(1,3,2)
plt.title('talk')
sns.set_context('talk', font_scale=1)

plt.subplot(1,3,3)
plt.title('poster')
sns.set_context('poster', font_scale=1)


plt.show()
````

Most of the plots, we can change the color palette used by `seaborn` for plotting. This is done by applying a different __`colormap`__ when we call the plotting function. For the available colormaps, see `colormap` documentation.

### 4.1 - Heapmaps

To create a heatmap, the DataFrame must be aligned such that the index column also is considered a 'column' like the normal column does. It no longer represent the unique indexing of each row. Let's see example:

````python

# 1 - Correlation
# The first way of creating a matrix is by creating a correlation matrix. Correlation basically shows how influential a variable
# is on the results

crash_matrix = crash_df.corr()

sns.heatmap( crash_matrix,
             annot = True,          # Show the value in middle of the rectangle
             cmap = 'Blues')
````

````python

# Another way: Pivot table

flights_df = sns.load_dataset('flights')

flights_df = flights_df.pivot_table( index = 'month',     # As we can see we've selected 3 columns to represent something
                               columns = 'year',          # Month on the row side, year on the column side, and values are
                               values= 'passengers')      # number of passengers

print( flights_df )

sns.heatmap( flights_df, cmap='Blues', 
             linecolor='white',
             linewidth=.1)
````

Cluster map can be complicated. It calculates the distances between 2 data, and try to join those data which are close to each other (having small differences between them) and group them into clusters. In other words, like-data are grouped together as much as possible.
Due to this behavior, Cluster map will rearrange our data columns and index so that like-data can be grouped together

````python
iris_df = sns.load_dataset('iris')

species = iris_df.pop('species')   # Just removing the column 'species'

sns.clustermap(iris_df)
````

### 4.4 - Pair Grid

Pair grid can be viewed as an extension to the pair plot. It allows us to define which plot shall be plotted and how the plot shall be plotted. Overall, it gives more control to us

````python

iris_df = sns.load_dataset('iris')

# Unlike previous plot which use functions, this time we create an object using Constructor
plt.figure( figsize=(10,20) )

iris_grid = sns.PairGrid( iris_df, hue='species' )

iris_grid.map_diag( plt.hist )               # Now we pass in functions, telling it the function to use to plot that subplot
iris_grid.map_offdiag(plt.scatter)






iris_grid = sns.PairGrid( iris_df, hue='species')

iris_grid.map_diag( plt.hist )
iris_grid.map_upper( plt.scatter )
iris_grid.map_lower( sns.kdeplot )

````

Facet grid enables us to print multiple plots in a grid, where we can define our grid and columns

````python

# In the constructor, we define the columns and rows of subplot grid
tips_fg = sns.FacetGrid( tips_df, col='time', row='smoker' )
tips_fg.map( plt.scatter, 'total_bill', 'tip' )
# Only here we define what function shall be used to plot. Also the row and column for the plotting function
````

````python

tips_fg = sns.FacetGrid( tips_df, col = 'time', hue='smoker',
                        height = 6,
                       aspect = 1,
                       col_order=['Dinner', 'Lunch'],
                       palette='Set1',
                       hue_kws = dict( marker = ['^', 'v'] ) )

tips_fg.map( plt.scatter, 'total_bill', 'tip', edgecolor='w')

plt.legend()
````
